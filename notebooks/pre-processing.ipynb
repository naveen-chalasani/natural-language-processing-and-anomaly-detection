```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('first_5_columns.csv', header=None, names=['date', 'time', 'id', 'type', 'component'])
df.head()
```

```python
df.info()
```

```python
df['id'].value_counts()
```

```python
df['type'].value_counts()
```

```python
df['row'] = np.arange(1, len(df) + 1)
df.head()
```

```python
y = pd.get_dummies(df['type'], prefix='type')
y.head()
```

```python
(sum(y['type_INFO']), sum(y['type_WARN']))
```

```python
df['type_INFO'] = y['type_INFO']
df['type_WARN'] = y['type_WARN']
df.drop('type', axis = 1, inplace = True)
df.head()
```

```python
(sum(df['type_INFO']), sum(df['type_WARN']))
```

```python
df['component'].value_counts()
```

```python
y = pd.get_dummies(df['component'], prefix='comp')
y.head()
```

```python
(sum(y['comp_dfs.FSNamesystem']), sum(y['comp_dfs.DataNode$PacketResponder']))
```

```python
df['comp_FSname'] = y['comp_dfs.FSNamesystem']
df['comp_PacketResponder'] = y['comp_dfs.DataNode$PacketResponder']
df['comp_DataXceiver'] = y['comp_dfs.DataNode$DataXceiver']
df['comp_FSDataset'] = y['comp_dfs.FSDataset']
df['comp_DataBlockScanner'] = y['comp_dfs.DataBlockScanner']
df['comp_DataNode'] = y['comp_dfs.DataNode']
df['comp_DataTransfer'] = y['comp_dfs.DataNode$DataTransfer']
df['comp_BlockReceiver'] = y['comp_dfs.DataNode$BlockReceiver']
df['comp_PendingMonitor'] = y['comp_dfs.PendingReplicationBlocks$PendingReplicationMonitor']

df.drop('component', axis = 1, inplace = True)
df.head()
```

```python
(sum(df['comp_FSname']), sum(df['comp_PacketResponder']))
```

```python
df['date'].value_counts()
```

```python
df.groupby('date').agg({'time': ['min', 'max']})
```

```python
df['time'].head()
```

```python
df['hour'] = pd.to_datetime(df['time'], format = '%H%M%S', errors='coerce').dt.hour
df['minute'] = pd.to_datetime(df['time'], format = '%H%M%S', errors='coerce').dt.minute
df['second'] = pd.to_datetime(df['time'], format = '%H%M%S', errors='coerce').dt.second

df.head()
```

```python
df['seconds'] = ((df['date'] - 81109) * 86400) + (df['hour'] * 3600) + (df['minute'] * 60) + df['second']

df.drop('date', axis = 1, inplace = True)
df.drop('time', axis = 1, inplace = True)
df.drop('hour', axis = 1, inplace = True)
df.drop('minute', axis = 1, inplace = True)
df.drop('second', axis = 1, inplace = True)

df.head()
```

```python
df['seconds'].fillna(method='ffill', inplace=True)
```

```python
df['seconds'].isnull().sum()
```

```python
df['minutes'] = (df['seconds'] // 60).astype('int')
df.drop('seconds', axis = 1, inplace = True)
df.drop('id', axis = 1, inplace = True)
df.head()
```

```python
df.info()
```

```python
df.to_csv('processed_01.csv', index=False, header=True)
```
