```python
df = pd.merge(df_event, df_block, how='inner', on='row', left_index=True)
df.head()
```

```python
event_seq = df.groupby(['block'])['template'].apply(' '.join).reset_index()
event_seq.head()
```

```python
event_seq['row'] = np.arange(1, len(event_seq) + 1)
event_seq.info()
```

# Add labels

```python
df_labels = pd.read_csv('anomaly_label.csv', header=0, names=['block', 'label'])
df_labels.head()
```

```python
df_labels.info()
```

```python
events_labels = pd.merge(event_seq, df_labels, how='inner', on='block', left_index=True)
events_labels.info()
```

```python
events_labels.head()
```

```python
events_labels.to_csv('events_labels_block.csv', index=False, header=True)
```
